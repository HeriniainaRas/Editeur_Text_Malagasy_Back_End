SCRAPING et création de dataset

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import urllib3
import re
import time

# Alphabet malagasy (à adapter si certaines lettres manquent)
alphabet = ["a","b","d","e","f","g","h","i","j","k","l","m","n","o","p","r","s","t","v","z"]

# Base URL pour lettres
base_url = "https://tenymalagasy.org/bins/rootLists?o=let{}"

# Autres liens spécifiques à scraper
extra_urls = [
    "https://tenymalagasy.org/bins/rootLists?o=posn",
    "https://tenymalagasy.org/bins/rootLists?o=pospn",
    "https://tenymalagasy.org/bins/rootLists?o=posaj",
    "https://tenymalagasy.org/bins/rootLists?o=posdv",
    "https://tenymalagasy.org/bins/rootLists?o=pospv",
    "https://tenymalagasy.org/bins/rootLists?o=posav",
    "https://tenymalagasy.org/bins/rootLists?o=posij",
    "https://tenymalagasy.org/bins/rootLists?o=posu"
]

# Désactiver warnings SSL
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# Liste pour stocker toutes les données
all_data = []

# Boucle sur chaque lettre de l'alphabet
for letter in alphabet:
    url = base_url.format(letter)
    print(f"Scraping {url} ...")

    try:
        response = requests.get(url, verify=False)
        html_content = response.text
        soup = BeautifulSoup(html_content, "lxml")

        table = soup.find("table", class_="menuLink")
        if not table:
            print(f" Aucun tableau trouvé pour {letter}")
            continue

        for row in table.find_all("tr"):
            cols = row.find_all("td")
            if len(cols) >= 2:
                racine = cols[0].get_text(strip=True)
                racine = re.sub(r"\(.*?\)", "", racine).strip()

                derivs_text = cols[1].get_text(strip=True)
                derivs_list = [d.strip() for d in derivs_text.split(",") if d.strip()]

                for mot in derivs_list:
                    all_data.append({"racine": racine, "mot": mot})

        time.sleep(1)  # pause pour ne pas surcharger le site

    except Exception as e:
        print(f" Erreur pour {letter}: {e}")

# Boucle sur les liens supplémentaires
for url in extra_urls:
    print(f"Scraping {url} ...")

    try:
        response = requests.get(url, verify=False)
        html_content = response.text
        soup = BeautifulSoup(html_content, "lxml")

        table = soup.find("table", class_="menuLink")
        if not table:
            print(f"Aucun tableau trouvé pour {url}")
            continue

        for row in table.find_all("tr"):
            cols = row.find_all("td")
            if len(cols) >= 2:
                racine = cols[0].get_text(strip=True)
                racine = re.sub(r"\(.*?\)", "", racine).strip()

                derivs_text = cols[1].get_text(strip=True)
                derivs_list = [d.strip() for d in derivs_text.split(",") if d.strip()]

                for mot in derivs_list:
                    all_data.append({"racine": racine, "mot": mot})

        time.sleep(1)

    except Exception as e:
        print(f"Erreur pour {url}: {e}")

#  Créer DataFrame unique
df = pd.DataFrame(all_data, columns=["racine", "mot"])
df.drop_duplicates(inplace=True)

#  Exporter CSV et Excel
df.to_csv("malagasy_dictionary_all.csv", index=False, encoding="utf-8")
df.to_excel("malagasy_dictionary_all.xlsx", index=False)

print(" Dataset complet créé avec succès !")
```
